## Creating subset of Products JSON to show true average value by SKU

As each entry has an individual "averge rating", I created a subset and took an overall average for each product identified by SKU

```python
products_json["sku"] = products_json["sku"].astype(str) 
#I was getting an error when running the code in cell 81, converting to a str fixed this
```

```python
#products_json.head()
```

```python
#Creating a more manageable subset to group the SKU's and get an overall mean for each

subset = products_json[["name_title","sku","average_product_rating","total_number_reviews"]]

#Removing empty values in sku column

subset = subset[subset["sku"] != "nan"]
```

```python
#display(subset.head())
print(subset.dtypes)
```

```python
# There are muptiple rows for each sku so I am grouping them together and caluclating the average of the
#"average_product_ rating" to get an average rating for each product

subset = subset.groupby(["name_title","sku"]).agg(
    total_av = ("average_product_rating","mean"), 
    total_reviews = ("total_number_reviews","sum"))
```

```python
#Sorting to see which products have the most/least reviews

subset_sorted = subset.sort_values(by = "total_reviews", ascending = False) #sorting values by descending order, total reviews to top
display(subset_sorted)
```

```python
print(sum(subset_sorted["total_reviews"] > 10))
# I recognised above that if I were to plot the subset that the data would be skewed by the products which only have 1 review.
# Therefore, I decided to only plot products which have more than 10 reviews
```

```python
# Creating a new sub set with products which have more than 10 reviews

x = subset_sorted["total_reviews"] > 10

top_reviewed = subset_sorted[x].reset_index() 

# Sorting by highest average review

top_reviewed_ordered = top_reviewed.sort_values(by = "total_av", ascending = False)

# Creating smaller subsets of the top and bottom 10 products by average review so that they fit on a bar chart

top_10 = top_reviewed_ordered.head(10)

bottom_10 = top_reviewed_ordered.tail(10)
```

```python
display(top_reviewed_ordered)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating a visualisation to look at the top and bottom 10 products by average user rating

```python
plt.figure(figsize=(12,6)) #Setting figure size

plt.subplot(121) #Defining sublot size and position
plt.title("Best 10 Products")
plt.ylabel("Average Customer Rating (Out of 5)")
plt.bar(top_10["name_title"],top_10["total_av"], 
        color="blue",
       edgecolor="black",
       linewidth=0.7) #color, edgecolor, linewidth to improve readbility

plt.xticks(ticks=range(len(top_10)),
           labels=top_10["name_title"],
           rotation=45,
          ha="right") #rotated x labels to improve readbility

plt.ylim(0,4.2)

plt.subplot(122)
plt.title("Worst 10 Products")
plt.bar(bottom_10["name_title"],bottom_10["total_av"],
       color="orangered",
       edgecolor="black",
       linewidth=0.7)

plt.xticks(ticks=range(len(top_10)),
           labels=bottom_10["name_title"],
           rotation=45,
          ha="right")

plt.ylim(0,4.2)

plt.suptitle("Best and Worst 10 Products by Average Customer Rating", fontsize=14)

plt.show()
```

```python
best_brands = brands_sorted.head(10).reset_index()
worst_brands = brands_sorted.tail(10).reset_index().sort_values(by = "total_reviews", ascending = False)
```

## Analysing number of reviews by age group

```python
reviews = pd.read_csv("reviews.csv")
print(reviews.shape)
display(reviews.head(5))
```

```python
users = pd.read_csv("users.csv")
print(users.shape)
display(users.head(5))
```

```python
merged = pd.merge(users,reviews, how="inner", on="Username") #Used inner join incase there was data in each set that doesn't match
```

```python
print(merged.shape)
```

```python
display(merged.head())
```

## Converting DOB's to age bands

```python
from datetime import datetime
```

```python
merged["DOB"] = pd.to_datetime(merged["DOB"], format="%d.%m.%Y", errors="coerce") #Ensuring DOB is right format

today = datetime.now()

merged["Age"] = 2025 - merged["DOB"].dt.year #Converting DOB to "age"
merged["Age"] = merged["Age"].astype(int, errors = "ignore") #Making "age" int values

bands = ["20-29", "30-39", "40-49", "50-59", "60-69", "70-79"] #Defining age bands
bins = [20,30,40,50,60,70,80] #Defining bins

merged["Age_band"] = pd.cut(merged["Age"], #Sorting values into bins
                              bins=bins,
                              labels=bands,
                             right=False)

ages_counts = merged.groupby("Age_band").size().reset_index(name = "Reviews") #Creating a subset to plot



display(merged.head())
display(ages_counts.head())
```

## Visualising data

```python
plt.figure(figsize=(7,7)) #Setting figure size

plt.bar(ages_counts["Age_band"],ages_counts["Reviews"], color="limegreen",edgecolor="black",
       linewidth=0.7) #Plotting bar chart

plt.suptitle("Number of Reviews by Age Group")
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Reviews", fontsize=12)
plt.ylim(3000,8500) #Changed y axis to better show differences between groups


plt.show()
```

For report - We can draw the conclusion that people in the xxx age groups leave reviews more often. We could also draw a rough conclusion that there
an increased number of sales with this group. However, this does not account for people who may have bought a product and not given a review. To accurately analyse sales by age group, we would require data on individual transactions and not just reviews.
